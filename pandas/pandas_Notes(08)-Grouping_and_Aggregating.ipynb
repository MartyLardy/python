## Grouping the data <code>.groupby()</code>
Split the object, Apply Function, Combine Results

```python
df["Country"].value_counts()
```

```python
df["LanguageWantToWorkWith"].value_counts()
```

### Lets split the object

```python
# group object (memory locatioin)
country_group = df.groupby(['Country'])
country_group
```

```python
# returns everyone from the United States
country_group.get_group("United States of America")
```

### the concept
#### Create a filter

```python
filt = (df["Country"] == "United States of America")
df.loc[filt]["WebframeHaveWorkedWith"].value_counts()
```

## Using the .groupby() with .value_counts() accessing df["something"]

### Why is it useful if i find it using the other filter method?

##### Lets play with the filter way of solving the problem

```python
filt = (df["Country"] == "India")
df.loc[filt]["LanguageWantToWorkWith"].str.contains("Python").sum()
```

```python
#This way is too complicated
country_group["LanguageWantToWorkWith"].apply(lambda hi: hi.str.contains("Python").sum())
```

```python
#figure what perceont pf people from each country know python?
```

```python
# solution:
#total number of respondence of each country
county_resp = df["Country"].value_counts()
```

```python
#total numober of rep who know python
county_knows_python = country_group["LanguageWantToWorkWith"].apply(lambda hi: hi.str.contains("Python").sum())
county_knows_python
```

#### New Method that combiles more than 1 series .concat([x,y, axis = "columns"])

```python
python_df = pd.concat([county_resp, county_knows_python],axis = "columns", sort = False)
python_df`
```

```python
python_df.rename(columns={"Country": "NumberRespons",
                          "LanguageWantToWorkWith": "RepKnowsPython"}
                 ,inplace = True)
python_df
```

```python
# create a new column:
python_df["PctKnowsPython"] = (county_knows_python/county_resp) * 100
# sort
python_df.sort_values(by="PctKnowsPython", ascending=False , inplace = True)
python_df.loc['Japan']
```
