### Drop missing values

```python
import pandas as pd
import numpy as np
```

```python
# we added extra data and null data (np.nan is NaN value)
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
# We will see this a lot. Find the missing data.
# Remove it! 
```

```python
df = pd.DataFrame(people)
```

```python
df
```

### Use .dropna()
#### dropna is using default arguments

```python
df.dropna()
```

```python
# defult argument for .dropna()
# if we set axis to columns, we tell pandas that we want to drop na in columns
# in index, we drop rows
df.dropna(axis='index', how = "any")
# if we use how = "all" , it will only drop rows with all rows with NA
```

```python
df.dropna(axis="index", how = "all")
# index 4 will be missing
```

```python
# one column will have at least one missing value
df.dropna(axis="columns", how = "any")
```

## We want to drop some missing values, but a specific column 
### We can use an subset argument

```python
df.dropna(axis = "index" , how = "any" , subset = ["email", "last"])
```

### Custom missing values
#### How do we handle it?

```python
df.replace("NA", np.nan, inplace = True)
df.replace("Missing", np.nan, inplace=True)
```

```python
df.dropna(axis="index", how="any", inplace = True)
df
```

### If we are unsure if it's a np.nan, we can run a mask with <code>.isna()</code>

```python
df.isna()
```

### Sometime we want to fill a na with another value <code>.fillna()</code>

```python
df = pd.DataFrame(people)

# mostly good for numerical values (0 or -1)
df.fillna("MISSING")
```

```python
df.fillna(0, inplace = True)
df
```

## Casting Data Types

```python
df = pd.DataFrame(people)
# all the columns is objects / string or a mix of things 
df.dtypes
```

```python
type(np.nan)
```

```python
df.replace('Missing', np.nan, inplace = True)
df['age'] = df['age'].astype(float)
df.isna()
```

#### Create a filter or a list of missing value key words

```python
na_vals = ["NA", "Missing"]
df2 = pd.read_csv("survey.csv", na_values=na_vals)
df2
```

### Casting the values

```python
df2["YearsCode"]

#conver it to a float:
df2["YearsCode"]
df2["YearsCode"].unique()
```

```python
# Casting = converting one value from one data type to another
# in this case, cast out ...."Less...More than..."
df2["YearsCode"].replace("Less than 1 year", 0, inplace = True)
df2["YearsCode"].replace("More than 50 years", 51, inplace = True)
df2["YearsCode"].astype(float).mean()
```
