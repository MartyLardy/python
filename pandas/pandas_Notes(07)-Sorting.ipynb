```python
people = {
    "first": ["corey","jane","john","adam"],
    "last": ["shafer","doe","doe","doe"],
    "email":["coreysharver@mail.com", "janedoe@email.com","jondoe@email.com","adamdoe@mail.com"],
}
```

```python
import pandas as pd
df = pd.DataFrame(people)
```

```python
df.head()
```

## Sorting  
### Using <code>.sort_values()</code>

```python
df.sort_values(by = "last")
```

### Decending and Ascending order

```python
#ascending
df.sort_values(by = "last" , ascending=True)
```

```python
# can sort more than one column
# first sort last, then sort first name
# decending last, accending first
df.sort_values(by = ["last","first"] , ascending=[False, True], inplace = True)
df
```

### Sorting Index<br>
<code>.sort_index()</code>

```python
# sort the index (back to the original
df.sort_index(ascending=True)
```

### Sorting a Series

```python
df['last'].sort_values()
```

## Sorting with Large Dataset

```python
pokemon_df = pd.read_csv("pokemon.csv")
```

```python
pokemon_df
```

## Sorting Multi Columns

```python
# lets sort by name and type ascending name, decendint type1
pokemon_df.sort_values(by = ["Name","Type 1"],
                       ascending=[True,False])
```

```python
# Lets look at name and Sp.Atk (ascending name, decending sp atk
pokemon_df.sort_values(by =["Name", "Sp. Atk"],
                       ascending=[True,False])
```

## Sorting Largest and Smallest Values

```python
# we want to see the 10 largest Sp. atk from a series
pokemon_df["Sp. Atk"].nlargest(10)
```

```python
# pass it into the complte dataframe

pokemon_df.nlargest(10, "Sp. Atk")
```

### Seeing the smallest

```python
pokemon_df.nsmallest(10, "Speed")
```
