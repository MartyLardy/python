## How to read and Write files:

### Lets look at .csv files first:
##### Read <code> read_csv("file name.csv") </code>
##### Write <code> to_csv("file_name.csv") </code>

```python
df = pd.read_csv("survey_results_public.csv", index_col = "ResponseId")
# set display (optional)
pd.set_option("display.max_columns" , 100)
pd.set_option("display.max_rows", 100)
df.head(3)
```

```python
# lets filter "Country"
filt = (df["Country"] == "United States of America") & (df["US_State"] == "New York")
len(df.loc[filt])
```

#### Lets save this filtered data to a new .csv file

```python
df_us_newyork = df.loc[filt]
df_us_newyork.to_csv("newyork_survey.csv")
```

### Tab delimited files <code>.tsv </code>

```python
#write
df_us_newyork.to_csv("newyork_survey.tsv" , sep = "\t")
```

```python
#read
nyc_tsv = pd.read_csv("newyork_survey.tsv", sep = "\t")
```
